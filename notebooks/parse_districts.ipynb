```python
districts_geojson_path = "../data/0301100100_UNITATS_ADM_POLIGONS.json"
```

```python
import json


with open(districts_geojson_path, 'r') as f:
    districts_geojson = json.load(f)
```

```python
districts = []
```

```python
output_path = "../data/districts.parquet"
```

```python
from shapely.geometry import shape
import geopandas as gpd

for feature in districts_geojson.get('features', []):
        properties = feature.get('properties', {})
        
        if properties.get('TIPUS_UA') == 'DISTRICTE':
            
            district_name = properties.get('NOM')
            
            geometry = shape(feature.get('geometry'))
            
            districts.append({
                'district_name': district_name,
                'geometry': geometry
            })
            
    if districts:
        gdf = gpd.GeoDataFrame(districts, geometry='geometry', crs="EPSG:4326")
        
        # Guardar la información en un archivo parquet
        # Se guarda como GeoParquet, estándar altamente compatible con pipelines espaciales
        gdf.to_parquet(output_parquet_path, index=False)
        print(f"Éxito: Se han guardado {len(gdf)} distritos en '{output_parquet_path}'.")
    else:
        print("No se encontró ningún feature con TIPUS_UA == 'DISTRICTE'.")
```
